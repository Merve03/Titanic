```python
def degiskenler(variable):
  cat=train_data[variable]
  sayi=cat.value_counts()
  plt.figure(figsize=(9,3))
  plt.bar(sayi.index,sayi)
  plt.xticks(sayi,index,sayi.index.value)
  plt.ylabel("frekans")
  plt.title(variable)
  plt.show()
  print("variabla , sayi")
    
```
